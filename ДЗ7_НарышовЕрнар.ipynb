```python
# кол-во пропущенных данных, например в TotalCharges не превышает 0,001%. Тут два варианта методов, это использовать медиану или же моду. 
# но мне захотелось выбрать mean()
#Однако важно отметить, что любой метод может внести систематическую ошибку в данные, особенно если отсутствующие данные отсутствуют случайно.
# но как говорится идеальности не существует.
missing_cols = df.columns[df.isna().any()]

for col in missing_cols:
    if col in df.columns:
        df[col].fillna(df[col].mean(), inplace=True)
```
